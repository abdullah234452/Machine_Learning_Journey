```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/placement.csv")
df.head(2)
```

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
upper_limit = df['placement_exam_marks'].quantile(0.95)
lower_limit_limit = df['placement_exam_marks'].quantile(0.05)
```

```python
#capping-winsorization
new_df_cap = df.copy()
```

```python
new_df_cap['placement_exam_marks'] = np.where(
    df['placement_exam_marks'] > upper_limit,
    upper_limit,
    np.where(
        df['placement_exam_marks'] < lower_limit_limit,
        lower_limit_limit,
        df['placement_exam_marks']
)
)
```

```python
sns.boxplot(new_df_cap['placement_exam_marks'])
```
